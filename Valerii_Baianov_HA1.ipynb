# 4. Creating new columns

```python
# Create `was_renovated` column. Bool column (0, 1) indicating whether the house was renovated.
df['was_renovated'] = df.yr_renovated.apply(lambda x: min(x, 1))
```

```python
from sklearn.preprocessing import KBinsDiscretizer

# Create new columns using the old ones (new column in your DataFrame)

# Q15.1 Create a `sqft_tot_area` column (sum of all columns with `sqft_` prefix) using any method above
sqft_col = df.columns[list(map(lambda x: x.find("sqft_") != -1, df.columns))].tolist()
df['sqft_tot_area'] = df[sqft_col].apply(sum, axis=1)

# Q15.2 Create a new column `sqm_tot_area` using `sqft_tot_area` and the fact that 1 foot = 0.3048 meters
df['sqm_tot_area'] = df.sqft_tot_area * 0.3048

# Q15.3 Create a new column `sqm_aver_floor_area` by dividing total area (in meters) by number of floors
df['sqm_aver_floor_area'] = df.sqm_tot_area / df.floors

# Q15.4 Create a new column `price_cat` by splitting a `price` into 5 ([1..5]) distinct intervals: 0 < x <=20%,
# 20% < x <= 40%, ... 80% < x <= 100% percentiles. You could use `.quantile()` to compute percentiles.
kbd = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df['price_cat'] = kbd.fit_transform(np.array(df.price).reshape(-1,1)) + 1

# Q15.5 Create a new bool column `high_class` it is True if the house has grade >= 9 and condition >= 4
df['high_class'] = (df.grade >= 9) & (df.condition >= 4)
```

# 6. Groupby

```python
# Create some groupby features

# Q19.1 `price_by_class` groupby `high_class` and compute median `price`.
df['high_class'] = df.high_class.apply(int)
df['price_by_class'] = df['high_class'].map(df.groupby(['high_class'])['price'].median())

# Q19.2 `price_by_year` groupby `year` and compute median price.
# Q19.3 `price_by_weekday` groupby `weekday` and compute median price.
# Q19.4 `area_by_price` groupby `price_cat` and compute average `sqft_living`.
df['area_by_price'] = df['price_cat'].map(df.groupby('price_cat').sqft_living.mean())

# Q19.5 `floors_by_age` groupby `floors` and compute average age of a house.
```

```python
# Create some other groupby features
# for this task check out this answer:
# https://stackoverflow.com/questions/47913343/how-to-groupby-and-map-by-two-columns-pandas-dataframe

# Q20.1 `n_houses_zipcode` groupby `zipcode` and count number of occurences of every unique zipcode
# Q20.2 `n_houses_yr_built` groupby `yr_built` and count number of houses built in each year
df1 = df.yr_built.to_frame().copy()
df1['id'] = df1.index
df['n_houses_yr_built'] = df['yr_built'].map(df1.groupby('yr_built').id.count())

# Q20.3 `price_by_yr_month_`(median, std) groupby `year`, `month` and compute median and std `price`.
# Q20.4 `price_by_grade_age_`(median, std) groupby `grade`, `house_age` and compute median and std `price`.
# Q20.5 `living_by_cond_`(median, std) groupby `waterfront`, `view`, `condition` and compute median and std `sqft_living`.
df_living_by_cond = df.groupby(['waterfront', 'view','condition']).sqft_living.aggregate(['median','std'])
df_living_by_cond.columns = ['living_by_cond_median', 'living_by_cond_std']
df = pd.merge(df, df_living_by_cond, on=['waterfront', 'view', 'condition'], how='left')
```

# 7. Building a regression model

```python
# Q21 Drop all generated features which used price column, e.g. price_by_year, price_cat.
df.drop(['price_cat', 'price_by_class', 'area_by_price'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.columns
```

```python
cat_features = ['zipcode', 'weekday', 'month', 'year']
drop_features = ['yr_renovated', 'lat', 'long', 'sqm_tot_area', 'sqft_tot_area']
```

```python
df.drop(drop_features, axis=1, inplace=True)
```

```python
df_cat = df[cat_features]
df_cat = pd.get_dummies(data=df_cat, columns=cat_features, drop_first=True)
df_cat.head()
```

```python
df = pd.concat([df[set(df.columns) - set(cat_features)], df_cat], axis=1)
df.head()
```

```python
# Q22 Split your data into train and test parts.
# How many records (rows) do you have in train and test tables? (list of int)?
from sklearn.model_selection import train_test_split

X, y = df.drop('price', axis=1), df.price
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=7)

[X_train.shape[0], X_test.shape[0]]

# Use sklearn.model_selection.train_test_split with test_size=0.33 and random_state=7
```

```python
# Create a predictive regression model of a house price.

# Q23.1 Use linear regression with l2 regularization (Ridge regression)
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler

ridge_pipe = make_pipeline(StandardScaler(), Ridge())
ridge_pipe.fit(X_train, y_train)

print("Ridge Train R2 = {}".format(ridge_pipe.score(X_train, y_train)))
print("Ridge Test R2 = {}".format(ridge_pipe.score(X_test, y_test)))

# Q23.2 Use decision tree regression
```

```python
# Q23.3 Use k nearest neighbours regression

knn_pipe = make_pipeline(StandardScaler(), KNeighborsRegressor())
knn_pipe.fit(X_train, y_train)

print("\nKNN Train R2 = {}".format(knn_pipe.score(X_train, y_train)))
print("KNN Test R2 = {}".format(knn_pipe.score(X_test, y_test)))
```

```python
# Use grid search to select optimal hyperparamters of your models. 

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import GridSearchCV

# Q24.1 Alpha for a ridge regression

ridge_pipe = make_pipeline(StandardScaler(), Ridge())
ridge_gs_cv = GridSearchCV(n_jobs=-1, verbose=10, estimator=ridge_pipe, param_grid={'ridge__alpha' : np.arange(1e-100, 1, 1e-2)})
ridge_gs_cv.fit(X_train, y_train)

best_alpha = ridge_gs_cv.best_params_['ridge__alpha']
print("Best Train Ridge R2 = {}, where alpha = {}".format(ridge_gs_cv.best_score_, best_alpha))

# Q24.2 Depth for the tree
# Q24.3 Number of neighbours for the knn
```

```python
# Q24.3 Number of neighbours for the knn

knn_pipe = make_pipeline(StandardScaler(), KNeighborsRegressor())
knn_gs_cv = GridSearchCV(n_jobs=-1, verbose=10, estimator=knn_pipe, param_grid={'kneighborsregressor__n_neighbors' : np.arange(1, 30)})
knn_gs_cv.fit(X_train, y_train)

best_n_neighbors = knn_gs_cv.best_params_['kneighborsregressor__n_neighbors']
print("Best Train KNN R2 = {}, where n_neighbors = {}".format(knn_gs_cv.best_score_, best_n_neighbors))
```

```python
# Compute train and test mean squared error for your best models (list of float).

# Q25.1 Train, test MSE using linear regression with l2 regularization
print("Best Train Ridge MSE = {}".format(np.around(mse(ridge_gs_cv.best_estimator_.predict(X_train), y_train), 3)))
print("Best Test Ridge MSE = {}\n".format(np.around(mse(ridge_gs_cv.best_estimator_.predict(X_test), y_test), 3)))

# Q25.2 Train, test MSE using decision tree regression

# Q25.3 Train, test MSE using k nearest neighbours regression
print("Best Train KNN MSE = {}".format(np.around(mse(knn_gs_cv.best_estimator_.predict(X_train), y_train), 3)))
print("Best Test KNN MSE = {}".format(np.around(mse(knn_gs_cv.best_estimator_.predict(X_test), y_test), 3)))
```

```python
# Compute train and test R^2 for your best models (list of float).

# Q26.1 Train, test R^2 using linear regression with l2 regularization

print("Best Ridge Train R2 = {}".format(np.around(ridge_gs_cv.best_estimator_.score(X_train, y_train), 3)))
print("Best Ridge Test R2 = {}\n".format(np.around(ridge_gs_cv.best_estimator_.score(X_test, y_test), 3)))

# Q26.2 Train, test R^2 using decision tree regression

# Q26.3 Train, test R^2 using k nearest neighbours regression
print("Best KNN Train R2 = {}".format(np.around(knn_gs_cv.best_estimator_.score(X_train, y_train), 3)))
print("Best KNN Test R2 = {}".format(np.around(knn_gs_cv.best_estimator_.score(X_test, y_test), 3)))
```
